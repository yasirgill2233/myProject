```python
import matplotlib.pylab as plt
import seaborn as sb

# Strength of students in each department
sb.countplot(x="Class", data=data, palette='Set2')
plt.title('Count of Students per Class')
plt.xlabel('Class')
plt.ylabel('Number of Students')
plt.show()
```

```python
sb.set(style="whitegrid")

# Visualization 1: Total fee collected by each class
plt.figure(figsize=(10, 6))
sb.barplot(x='Class', y='Fee', data=data, estimator=sum)
plt.title('Total Fee Collected by Each Class')
plt.xlabel('Class')
plt.ylabel('Total Fee')
plt.show()
```

```python
# Check how many students pay same fee
import matplotlib.pylab as plt
import seaborn as sb
plt.figure(figsize=(10, 6))
sb.countplot(x="Fee", data=data, palette='tab10')

plt.title('No of students that pay same fee')
plt.xlabel('Fee')
plt.ylabel('Number of Students')
plt.show()
```

```python
# Visualization 2: Scatter plot of Roll.No vs Fee
plt.figure(figsize=(10, 6))
sb.scatterplot(x='Roll.No', y='Fee', data=data, hue='Class', palette='deep')
plt.title('Scatter Plot of Roll.No vs Fee')
plt.xlabel('Roll.No')
plt.ylabel('Fee')
plt.legend(title='Class')
plt.show()
```

```python
# Visualization 5: Overall Fee Distribution
plt.figure(figsize=(10, 6))
sb.histplot(data['Fee'], bins=10, kde=True)
plt.title('Overall Fee Distribution')
plt.xlabel('Fee')
plt.ylabel('Frequency')
plt.show()
```
